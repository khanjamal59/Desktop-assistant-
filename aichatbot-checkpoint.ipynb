```python
import pyttsx3
import speech_recognition as sr
import webbrowser  
import datetime  
import wikipedia 
from AppOpener import open
import os

def take_command():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print('Listening...')
        r.pause_threshold = 0.7
        try:
            audio = r.listen(source)
            query = r.recognize_google(audio, language='en-in')
            print(f"Recognized: {query}")
            return query.lower()
        except sr.UnknownValueError:
            print("Sorry, I did not understand that.")
            return None
        except sr.RequestError as e:
            print(f"Error with the speech recognition service: {e}")
            return None

def speak(audio):
    engine = pyttsx3.init()
    voices = engine.getProperty('voices')
    engine.setProperty('voice', voices[0].id)
    engine.say(audio)
    engine.runAndWait()

def tell_day():
    day = datetime.datetime.today().weekday() + 1
    Day_dict = {1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday', 7: 'Sunday'}
    day_of_the_week = Day_dict.get(day, 'Unknown day')
    print(day_of_the_week)
    speak(f"Today is {day_of_the_week}")

def tell_time():
    now = datetime.datetime.now()
    hour = now.strftime("%H")
    minute = now.strftime("%M")
    speak(f"The time is {hour} hours and {minute} minutes")

def hello():
    speak("Hello sir, I am your desktop assistant. How may I help you today?")

def take_query():
    hello()
    while True:
        query = take_command()
        if query is None:
            continue

        if "open youtube" in query:
            speak("Opening YouTube")
            webbrowser.open("http://www.youtube.com")
        elif "open google" in query:
            speak("Opening Google")
            webbrowser.open("http://www.google.com")
        elif "which day it is" in query:
            tell_day()
        elif "tell me the time" in query:
            tell_time()
        elif "ok bye" in query or "exit" in query:
            speak("Goodbye! Have a nice day.")
            break
        elif "open calculator" in query:
            speak("Opening Calculator")
            open("calculator")
        elif "open whatsapp" in query:
            speak("Opening WhatsApp")
            open("whatsapp")
        elif "from wikipedia" in query:
            query = query.replace("wikipedia", "").strip()
            if query:
                speak("Searching Wikipedia")
                try:
                    result = wikipedia.summary(query, sentences=4)
                    speak("According to Wikipedia")
                    speak(result)
                except wikipedia.exceptions.DisambiguationError as e:
                    speak(f"Multiple results found: {e.options}")
                except wikipedia.exceptions.PageError:
                    speak("No page found for the query.")
        elif "tell me about yourself" in query:
            speak("I am a desktop assistant created using Python.")
        elif "who made you" in query:
            speak("I was created by Jamaal.")

if __name__ == '__main__':
    take_query()
```
